```python
#  1  -  Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias 
# de cada letra en la cadena. Los espacios no deben ser considerados.

def contar_frecuencia_letras (cadena_texto):
    
    """Función que recibe una cadena de texto y la devuelve como diccionario con las fecuencias de cada letra de la cadena
    
    Args:
        string : cadena de texto

    Return:
        diccionario : frecuencia de cada letra
    """
    cadena_texto = cadena_texto.replace (" ","")
    frecuencia = {}
    for letra in cadena_texto:
        if letra in frecuencia:
            frecuencia[letra] += 1
        else:
            frecuencia[letra] = 1

    return frecuencia
    
def contar_frecuencia_letras_non_case_sensitive (cadena_texto):
    cadena_texto = cadena_texto.lower()
    
    return contar_frecuencia_letras(cadena_texto)

print(contar_frecuencia_letras_non_case_sensitive('Esternocleidomastoideo'))
```

```python
# 29 - Crea una función que convierta una variable en una cadena de texto y enmascare todos los caracteres  con el 
# carácter '#', excepto los últimos cuatro.

def cadena_texto_especial (variable):
    """funcion que convierte una variable en una cadena de texto y enmascara todos los caracteres con el caracter '#', excepto los ultimos 4
    Args:
        str,int,float : una cadena de texto
    Retuen:
        str : Cadena de texto enmascarada con '#', menos los ultimos 4 digitos
    """
    texto = str(variable)
    if len(texto) <= 4:
        return texto
    else:
        return '#' * (len(texto) - 4) + texto [-4:]
        

print (cadena_texto_especial('76805674'))
print(cadena_texto_especial("Hola mundo"))
print(cadena_texto_especial("1234")) 

```

```python
# 30 - Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras pero en diferente orden.

from collections import Counter

def anagrama (palabra1,palabra2):
    """Funcion que determina si dos palabras son anagramas
    Args:
        str : un palabra
    Resturn:
        str : Si es un anagrama o no
    """


    if len(palabra1) != len(palabra2):
        return 'No es un anagrama, no tiene el mimso numero de letras'
    else:
        if Counter(palabra1) == Counter(palabra2):
            return 'Si que son anagramas'
        else:
            return 'Tienen la misma longitud, pero no son anagramas'
    
print(anagrama("lorc","calor"))
```

```python
# 36 - Crea la clase UsuarioBanco ,representa a un usuario de un banco con su nombre, saldo y si tiene o no cuenta corriente. 
# Proporciona métodos para realizar operaciones como retirar dinero, transferir dinero desde otro usuario y agregar dinero al saldo.
# Código a seguir:
# 1 Inicializar un usuario con su nombre, saldo y si tiene o no cuenta corriente mediante 
# 2 Implementar el método True y False .retirar_dinero para retirar dinero del saldo del usuario. Lanzará un error en caso de no poder hacerse.
# 3 Implementar el método transferir_dinero para realizar una transferencia desde otro usuario al usuario actual. Lanzará un error en caso de no poder hacerse.
# 4 Implementar el método agregar_dinero para agregar dinero al saldo del usuario.
# Caso de uso:
# 1 Crear dos usuarios: "Alicia" con saldo inicial de 100 y "Bob" con saldo inicial de 50, ambos con cuenta corriente.
# 2 Agregar 20 unidades de saldo de "Bob".
# 3 Hacer una transferencia de 80 unidades desde "Bob" a "Alicia".
# 4 Retirar 50 unidades de saldo a "Alicia".

class UsuarioBanco:
    def __init__(self,nombre, saldo, cuenta_corriente):
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente
    def retirar_dinero(self,cantidad):
        if not self.cuenta_corriente:
            raise ValueError("El usuario no tiene cuenta corriente")
        if self.saldo < cantidad:
            raise ValueError("No hay dinero suficiente")
        else:
            self.saldo -= cantidad
    def transferir_dinero(self,otro_usuario,cantidad):
        otro_usuario.retirar_dinero(cantidad)
        self.agregar_dinero(cantidad)

    def agregar_dinero(self,cantidad):
        self.saldo += cantidad
# Caso de uso:
try:
    alicia = UsuarioBanco("alicia",100,True)
    bob = UsuarioBanco("bob",50,True)

    bob.agregar_dinero(20)
    alicia.transferir_dinero(bob,80)
    alicia.retirar_dinero(50)
except ValueError as e:
    print("Error:",e)
print(f"El saldo de alicia es:" ,alicia.saldo)
print(f"El saldo de bob es:" ,bob.saldo)
```

```python
# 37 - Crea una función llamada procesar_texto que procesa un texto según la opción especificada: reemplazar_palabras, contar_palabras , eliminar_palabra. 
# Estas opciones son otras funciones que tenemos que definir primero y llamar dentro de la función procesar_texto .
# Código a seguir:
# 1 Crear una función contar_palabras para contar el número de veces que aparece cada palabra en el texto. Tiene que devolver un diccionario.
# 2 Crear una función reemplazar_palabras para remplazar una palabra_original del texto por una palabra_nueva. Tiene que devolver el texto con el remplazo de palabras.
# 3 Crear una función eliminar_palabra para eliminar una palabra del texto. Tiene que devolver el texto con la palabra eliminada.
# 4 Crear la función procesar_texto que tome un texto, una opción(entre "contar", "reemplazar", "eliminar") y un número de argumentos variable según la opción indicada.
# Caso de uso:
# Comprueba el funcionamiento completo de la función procesar_text

def contar_palabras(texto):
    palabras = texto.split()
    numero_palabras = {}
    for palabra in palabras:
        if palabra in numero_palabras:
            numero_palabras[palabra] += 1
        else:
            numero_palabras[palabra] = 1
    return numero_palabras
texto = "tres tristes tigres comen trigo en un trigal , un tigre , dos tigres , tres tigres"

print(contar_palabras(texto))

def reemplazar_palabras (texto, palabra_original, palabra_nueva):
    texto_modificado = texto.replace(palabra_original, palabra_nueva)
    return texto_modificado
texto = "me gusta la fruta"
print(reemplazar_palabras(texto,"fruta","platano"))

def eliminar_palabra (texto, palabra):
   palabras = texto.split()
   palabras_filtradas = [p for p in palabras if p != palabra]
   texto_resultante = " ".join(palabras_filtradas)
   return texto_resultante
texto = "me gusta la fruta"
print(eliminar_palabra(texto,"la"))

def procesar_texto(texto,opcion,*args):
    if opcion == "contar":
        return contar_palabras(texto)
    elif opcion == "reemplazar":
        return reemplazar_palabras(texto, args[0], args[1])
    elif opcion == "eliminar":
        return eliminar_palabra(texto, args[0])
    else:
        raise ValueError ("Opcion no valida")
    
texto = "A Ana le gusta la fruta y la verdura"
print(procesar_texto(texto,"contar"))
print(procesar_texto(texto,"reemplazar","Ana","Carlos"))
print(procesar_texto(texto,"eliminar","verdura"))
    

```

```python
# 38 - Genera un programa que nos diga si es de noche, de día o tarde según la hora proporcionada por el usuario.



def definir_hora(hora):
    if hora >= 9 and hora <= 13:
        return "día"
    elif hora > 13 and hora <= 21:
        return "tarde"
    else:
        return "noche"
hora = int(input("Introduce la hora (0-23):"))
print(definir_hora(hora))
```

```python
# 41 - En este ejercicio, se te pedirá que escribas un programa en Python que utilice condicionales para determinar el 
# monto final de una compra en una tienda en línea, después de aplicar un descuento. El programa debe hacer lo 
# siguiente:
# 1. Solicita al usuario que ingrese el precio original de un artículo.
# 2. Pregunta al usuario si tiene un cupón de descuento (respuesta sí o no).
# 3. Si el usuario responde que sí, solicita que ingrese el valor del cupón de descuento.
# 4. Aplica el descuento al precio original del artículo, siempre y cuando el valor del cupón sea válido (es decir, mayor 
# a cero). Por ejemplo, descuento de 15€. 
# 5. Muestra el precio final de la compra, teniendo en cuenta el descuento aplicado o sin él. 
# 6. Recuerda utilizar estructuras de control de flujo como if, elif y else para llevar a cabo estas acciones en tu 
# programa de Python. 

precio = float(input("Ingrese el precio originario del articulo"))
tiene_cupon = input("¿Tienes cupon de descuento?(resupuesta si o no)" ).lower()

if tiene_cupon == "si":
    descuento = float(input("Ingrese el valor del cupon de descuento"))
    if descuento > 0 and descuento <= precio:
        precio_final = precio - descuento
    else:
        print("El cupon no es valido")
        precio_final = precio
elif tiene_cupon == "no":
    precio_final = precio
else:
    print("Respuesta no valida")
    precio_final = precio
print("El precio final es:" ,precio_final)

```
